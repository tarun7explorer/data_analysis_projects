Here I'm generating ta all possible plots with MatPlot Library!!

Bar Plot

```python
# Bar Plot
import matplotlib.pyplot as plt

# Convert 'Date/Time' to datetime objects
data['Date/Time'] = pd.to_datetime(data['Date/Time'])

# Extract month and calculate the average temperature for each month
monthly_avg = data.groupby(data['Date/Time'].dt.month)['Temp_C'].mean()

monthly_avg.plot(kind='bar')
plt.title('Average Monthly Temperature')
plt.xlabel('Month')
plt.ylabel('Temperature')
plt.show()
```

Pie Chart

```python
# PieChart
weather_counts = data['Weather'].value_counts()
top_n = 10
weather_counts_top = weather_counts.head(top_n)

plt.figure(figsize=(10, 10))
plt.pie(weather_counts_top, labels=weather_counts_top.index, autopct='%1.1f%%', startangle=140)
plt.title(f'Top {top_n} Weather Conditions Distribution')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

Line plot of temperature over time

```python
plt.figure(figsize=(12, 6))
plt.plot(data['Date/Time'], data['Temp_C'])
plt.title('Temperature over Time')
plt.xlabel('Date/Time')
plt.ylabel('Temperature (Â°C)')
plt.show()
```

Scatter plot of Temperature vs Humidity

```python
plt.figure(figsize=(10, 6))
plt.scatter(data['Temp_C'], data['Rel Hum_%'])
plt.title('Temperature vs. Relative Humidity')
plt.xlabel('Temperature (Â°C)')
plt.ylabel('Relative Humidity (%)')
plt.show()
```

Bar plot of the top 10 most frequent weather conditions

```python
weather_counts = data['Weather'].value_counts().head(10)
plt.figure(figsize=(12, 6))
weather_counts.plot(kind='bar')
plt.title('Top 10 Most Frequent Weather Conditions')
plt.xlabel('Weather Condition')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

Box plot of Wind Speed

```python
plt.figure(figsize=(8, 6))
data.boxplot(column='Wind Speed_km/h')
plt.title('Box Plot of Wind Speed')
plt.ylabel('Wind Speed (km/h)')
plt.show()
```

Histogram of Pressure

```python
plt.figure(figsize=(10, 6))
plt.hist(data['Press_kPa'], bins=20)
plt.title('Histogram of Pressure')
plt.xlabel('Pressure (kPa)')
plt.ylabel('Frequency')
plt.show()
```

Now I'm generating all possible plots & graphs using Seabron Library

import the seaborn library.

```python
import seaborn as sns
```


Scatter plot using Seaborn to visualize the relationship between Temperature and Relative Humidity:

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Temp_C', y='Rel Hum_%', data=data)
plt.title('Temperature vs. Relative Humidity (Seaborn)')
plt.xlabel('Temperature (Â°C)')
plt.ylabel('Relative Humidity (%)')
plt.show()
```

Histogram of Wind Speed to show the distribution:

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['Wind Speed_km/h'], bins=20, kde=True)
plt.title('Distribution of Wind Speed')
plt.xlabel('Wind Speed (km/h)')
plt.ylabel('Frequency')
plt.show()
```

Box plot of Temperature by Month using Seaborn to see the temperature distribution throughout the year:

```python
# Ensure the Date/Time column is in datetime format
data['Date/Time'] = pd.to_datetime(data['Date/Time'])
data['Month'] = data['Date/Time'].dt.month

plt.figure(figsize=(12, 6))
sns.boxplot(x='Month', y='Temp_C', data=data)
plt.title('Temperature Distribution by Month')
plt.xlabel('Month')
plt.ylabel('Temperature (Â°C)')
plt.show()
```

Count plot of the top 10 most frequent Weather Conditions

```python
plt.figure(figsize=(12, 6))
sns.countplot(y='Weather', data=data, order=data['Weather'].value_counts().nlargest(10).index)
plt.title('Top 10 Most Frequent Weather Conditions (Seaborn)')
plt.xlabel('Frequency')
plt.ylabel('Weather Condition')
plt.show()
```

Correlation heatmap of the numerical columns in the dataset

```python
plt.figure(figsize=(10, 8))
sns.heatmap(data.select_dtypes(include=['number']).corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
```
